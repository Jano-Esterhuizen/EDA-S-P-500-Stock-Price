```python
import pandas as pd

# URL of the Wikipedia page
url = "https://en.wikipedia.org/wiki/List_of_S%26P_500_companies"

# Read all tables from the webpage
tables = pd.read_html(url)

# The first table on the page contains the list of S&P 500 companies
df = tables[0]

df
```

# Examining The Sectors

```python
sector_unique = df["GICS Sector"].unique()

print("List of unique Sectors:\n" + "\n".join(sector_unique))
print("number of unique Sectors " + str(len(sector_unique)))
```

## Plot Closing Price of Query Symbol

```python
import matplotlib.pyplot as plt

def price_plot(symbol):
    df = pd.DataFrame(data[symbol].Close)
    df['Date'] = df.index
    plt.fill_between(df.Date, df.Close, color="skyblue", alpha=0.3)
    plt.plot(df.Date, df.Close, color="skyblue", alpha=0.8)
    plt.xticks(rotation=90)
    plt.title(symbol, fontweight="bold")
    plt.xlabel("Date", fontweight="bold")
    plt.ylabel("Closing Price", fontweight="bold")
    return plt.show()
```

```python
price_plot("AAPL")
```

```python
for i in list(df["Symbol"])[:10]:
    price_plot(i)
```
